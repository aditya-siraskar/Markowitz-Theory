### Step 1 Data Aquisition and Importing 

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style configuration for "Stunning" dark plots
plt.style.use('dark_background')

# 1. Define Tickers and Timeframe
tickers = ['AAPL', 'JPM', 'JNJ', 'NKE'] 
start_date = '2020-01-01'
end_date = '2025-01-01'

# 2. Download Data
# FIX: We add 'auto_adjust=True' and select 'Close'.
# This ensures we get the split/dividend adjusted price, which is critical for accurate returns.
print(f"Downloading data for {tickers}...")
data = yf.download(tickers, start=start_date, end=end_date, auto_adjust=True)['Close']

# Check if data downloaded correctly
if data.empty:
    print("Error: No data downloaded. Check your internet connection or ticker symbols.")
else:
    print("Data downloaded successfully!")

# 3. Calculate Daily Log Returns
log_returns = np.log(data / data.shift(1)).dropna()

# 4. Annualize the Data
# We multiply by 252 (trading days in a year) to scale daily metrics to annual
annual_mean_returns = log_returns.mean() * 252
annual_cov_matrix = log_returns.cov() * 252

# --- VISUALIZATION 1: CORRELATION HEATMAP ---
plt.figure(figsize=(10, 8))
sns.heatmap(log_returns.corr(), annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Asset Correlation Matrix', fontsize=16, color='white')
plt.show()
```

# ðŸ“š Quick Theory Revision: Data Preparation

Before running the Markowitz simulation, we processed the raw data. Here is *why* we did what we did:

### 1. Why Log Returns? ($R_t$)
Instead of simple percentage changes, we calculated Logarithmic Returns:
$$R_t = \ln\left(\frac{P_t}{P_{t-1}}\right)$$
* **Time Additivity:** Log returns can be summed over time (Simple returns cannot). This makes calculating multi-period returns cleaner.
* **Normality Assumption:** Log returns are theoretically unbounded ($-\infty$ to $+\infty$), fitting the "Normal Distribution" assumption of the Markowitz model better than simple returns (which are bounded at -100%).

### 2. Why Correlation vs. Covariance?
* **The Math uses Covariance:** The optimization algorithm needs *Covariance* to calculate the portfolio variance (Risk).
* **The Human uses Correlation:** We plotted the *Correlation Matrix* because it standardizes relationships between **-1 and 1**. This allows us to visually spot diversification opportunities before running the code.

### 3. Why Multiply by 252? (Annualization)
The raw data is **daily**. Investors think in **years**.
* There are $\approx$ **252 trading days** in a year.
* **Returns:** We multiply mean daily returns by 252 to get *Annualized Return*.
* **Covariance:** We multiply the daily covariance matrix by 252 to get *

### Step 2: The Monte Carlo Simulation (The "Cloud")

```python
num_portfolios = 5000
all_weights = np.zeros((num_portfolios, len(tickers)))
ret_arr = np.zeros(num_portfolios)
vol_arr = np.zeros(num_portfolios)
sharpe_arr = np.zeros(num_portfolios)

for i in range(num_portfolios):
    # Create random weights
    weights = np.array(np.random.random(len(tickers)))
    weights = weights / np.sum(weights)  # Normalize so sum is 1
    
    # Record weights
    all_weights[i, :] = weights
    
    # Expected Return
    ret_arr[i] = np.sum(annual_mean_returns * weights)
    
    # Expected Volatility (Risk) -> The "Std Dev" Formula
    vol_arr[i] = np.sqrt(np.dot(weights.T, np.dot(annual_cov_matrix, weights)))
    
    # Sharpe Ratio (Return / Risk)
    sharpe_arr[i] = ret_arr[i] / vol_arr[i]

# Store data in a DataFrame for easy plotting
sim_data = pd.DataFrame({
    'Return': ret_arr,
    'Volatility': vol_arr,
    'Sharpe': sharpe_arr
})

print("Simulation complete.")

# Check the first 5 generated portfolios
print(sim_data.head())
```

# ðŸ“š Quick Theory Revision: Monte Carlo Simulation

In this step, we used a **Brute Force** approach to visualize the trade-off between Risk and Return.

### 1. What is a Monte Carlo Simulation?
Instead of solving an equation immediately, we generated **5,000 random portfolios**. By plotting thousands of random possibilities, we can see the "shape" of the marketâ€”what is possible and what is not.

### 2. The Workflow inside the Loop
1.  **Random Weights:** Assign random proportions to each asset.
2.  **Normalization:** Scale weights so they sum to exactly 1.0 (100% of capital).
3.  **Calculate Metrics:** Compute the Expected Return and Volatility (Standard Deviation) for that specific random combination.

### 3. The Matrix Calculation for Risk
We used Linear Algebra to calculate portfolio risk:
$$\sigma_p = \sqrt{w^T \cdot \Sigma \cdot w}$$
* $w$: The weights of the assets.
* $\Sigma$ (Sigma): The Annualized Covariance Matrix.
* **Why?** This formula captures not just individual asset risks, but also how they interact (Correlation). This is where diversification is mathematically calculated.

### 4. The Goal
The resulting "cloud" of points allows us to visually identify the **Efficient Frontier**â€”the upper boundary of the cloud where returns are maximized for a given level of risk.

### Step 3: Finding the Optimal Frontier (The "Line")

```python
# Locate the "Best" Portfolios from our simulation
max_sharpe_idx = sim_data['Sharpe'].idxmax()
min_vol_idx = sim_data['Volatility'].idxmin()

max_sharpe_port = sim_data.iloc[max_sharpe_idx]
min_vol_port = sim_data.iloc[min_vol_idx]

# --- VISUALIZATION 2: THE EFFICIENT FRONTIER ---
plt.figure(figsize=(12, 8))

# 1. Plot the "Cloud" of portfolios
plt.scatter(sim_data['Volatility'], sim_data['Return'], c=sim_data['Sharpe'], cmap='plasma', alpha=0.6, s=10)
plt.colorbar(label='Sharpe Ratio')

# 2. Mark the Max Sharpe Ratio (Red Star)
plt.scatter(max_sharpe_port['Volatility'], max_sharpe_port['Return'], c='red', s=200, marker='*', edgecolors='white', label=f'Max Sharpe ({max_sharpe_port["Sharpe"]:.2f})')

# 3. Mark the Min Volatility (Blue Star)
plt.scatter(min_vol_port['Volatility'], min_vol_port['Return'], c='blue', s=200, marker='*', edgecolors='white', label=f'Min Volatility ({min_vol_port["Volatility"]:.2f})')

# Labels and Titles
plt.xlabel('Expected Volatility (Risk)', fontsize=14)
plt.ylabel('Expected Return', fontsize=14)
plt.title('Markowitz Efficient Frontier: Monte Carlo Simulation', fontsize=18)
plt.legend(labelcolor='white', loc='upper left') # Legend color fix for dark mode
plt.grid(True, linestyle='--', alpha=0.3)

plt.show()

# Print the weights for the optimal portfolio
print("-" * 30)
print("Optimal Portfolio Weights (Max Sharpe):")
print("-" * 30)
optimal_weights = all_weights[max_sharpe_idx]
for ticker, weight in zip(tickers, optimal_weights):
    print(f"{ticker}: {weight*100:.2f}%")
```

# ðŸ“š Quick Theory Revision: The Efficient Frontier

In this final step, we visualized the results of our simulation to identify the "Best" portfolios.

### 1. Identifying the "Stars"
We used Pandas to find two specific portfolios from our 5,000 random samples:
* **The Red Star (Max Sharpe Ratio):**
    * *Found via:* `sim_data['Sharpe'].idxmax()`
    * *Meaning:* This is the portfolio with the **highest Risk-Adjusted Return**. It is the most "efficient" use of your capital.
* **The Blue Star (Min Volatility):**
    * *Found via:* `sim_data['Volatility'].idxmin()`
    * *Meaning:* This is the portfolio with the **lowest possible risk**. It is the "safest" bet using these specific assets.

### 2. Reading the Graph
The Scatter Plot reveals the geometry of the market:
* **The "Cloud":** Represents the universe of possible portfolios.
* **The Colors:** Brighter/Warmer colors indicate a higher Sharpe Ratio (better performance).
* **The Upper-Left Edge:** This is the **Efficient Frontier**.
    * *Definition:* The set of optimal portfolios that offer the highest expected return for a defined level of risk.
    * *Key Insight:* You cannot find a portfolio *above* this line.

### 3. Investment Zones
* **ðŸš« The "Impossible" Zone (Top Left):** High returns with zero risk do not exist.
* **ðŸ“‰ The "Inefficient" Zone (Bottom Right):** Portfolios here have high risk but low returns. A rational investor would never choose these; they would move vertically up to the Efficient Frontier instead.

### Step 3.5: Extracting the Optimal Weights

```python
# Create a function to print weights nicely
def print_portfolio_weights(weights, title):
    print(f"--- {title} ---")
    for ticker, weight in zip(tickers, weights):
        print(f"{ticker}: {weight*100:.2f}%")
    print("-" * 20)

# 1. Extract weights for the Max Sharpe Portfolio (Red Star)
max_sharpe_weights = all_weights[max_sharpe_idx]

# 2. Extract weights for the Min Volatility Portfolio (Blue Star)
min_vol_weights = all_weights[min_vol_idx]

# 3. Print the Report
print_portfolio_weights(max_sharpe_weights, "Max Sharpe Ratio Portfolio (Growth)")
print_portfolio_weights(min_vol_weights, "Min Volatility Portfolio (Safety)")

# 4. Compare Metrics
print("\n--- Performance Comparison ---")
print(f"Max Sharpe -> Return: {sim_data.iloc[max_sharpe_idx]['Return']:.2%}, Risk: {sim_data.iloc[max_sharpe_idx]['Volatility']:.2%}")
print(f"Min Volatility -> Return: {sim_data.iloc[min_vol_idx]['Return']:.2%}, Risk: {sim_data.iloc[min_vol_idx]['Volatility']:.2%}")
```

# ðŸ“š Quick Theory Revision: Portfolio Allocation

After identifying the optimal portfolios on the graph, we extracted the specific **Asset Allocation**.

### 1. Decoding the "Max Sharpe" Weights
This allocation answers the question: *"How do I structure my portfolio to get the best possible return per unit of risk?"*
* It often favors assets with high returns and low correlations to others.

### 2. Decoding the "Min Volatility" Weights
This allocation answers the question: *"How do I structure my portfolio to lose the least amount of sleep?"*
* It heavily favors stable assets (low standard deviation) and those that hedge each other (negative correlation).
* You will often see "riskier" high-growth stocks vanish from this list.

### Step 4: Mathematical Optimization.

```python
from scipy.optimize import minimize

# 1. Define the Objective Function (Minimize Variance)
def minimize_volatility(weights):
    return np.sqrt(np.dot(weights.T, np.dot(annual_cov_matrix, weights)))

# 2. Define Constraints and Bounds
# Constraint 1: Weights must sum to 1
cons_sum_weights = {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
# Bounds: 0 to 1 for each asset (No short selling)
bounds = tuple((0, 1) for _ in range(len(tickers)))

# 3. Calculate the Efficient Frontier Line
# We will find the lowest risk for a range of target returns
target_returns = np.linspace(sim_data['Return'].min(), sim_data['Return'].max(), 50)
frontier_volatility = []

for target in target_returns:
    # Constraint 2: Portfolio return must equal the target return
    cons_target_return = {'type': 'eq', 'fun': lambda x: np.sum(x * annual_mean_returns) - target}
    
    # Run the optimization
    constraints = [cons_sum_weights, cons_target_return]
    result = minimize(minimize_volatility, 
                      x0=len(tickers)*[1./len(tickers)], # Start with equal weights
                      bounds=bounds, 
                      constraints=constraints,
                      method='SLSQP')
    
    frontier_volatility.append(result['fun'])

# 4. Plotting the Result
plt.figure(figsize=(12, 8))

# Re-plot the Monte Carlo Simulation (The "Cloud") for context
plt.scatter(sim_data['Volatility'], sim_data['Return'], c=sim_data['Sharpe'], cmap='plasma', alpha=0.3, s=10)

# Plot the Mathematical Efficient Frontier (The "Line")
plt.plot(frontier_volatility, target_returns, 'w--', linewidth=3, label='Mathematical Efficient Frontier')

# Re-plot the Stars from Step 3
plt.scatter(max_sharpe_port['Volatility'], max_sharpe_port['Return'], c='red', s=200, marker='*', edgecolors='white', label='Max Sharpe (Simulated)')
plt.scatter(min_vol_port['Volatility'], min_vol_port['Return'], c='blue', s=200, marker='*', edgecolors='white', label='Min Volatility (Simulated)')

plt.xlabel('Expected Volatility (Risk)')
plt.ylabel('Expected Return')
plt.title('The Efficient Frontier: Simulation vs. Math')
plt.legend(labelcolor='white')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

# ðŸ“š Quick Theory Revision: Mathematical Optimization

In this final step, we moved from "guessing" (Monte Carlo) to "solving" using Calculus.

### 1. Why `scipy.optimize`?
While the Monte Carlo simulation gave us a good approximation (the cloud), it didn't give us the *exact* boundary. To find the perfect Efficient Frontier, we used **Sequential Least Squares Programming (SLSQP)**.

### 2. The Logic of the Loop
We didn't just solve one equation. We ran a loop:
1.  **Set a Target Return:** "I want exactly 15% return."
2.  **Minimize Volatility:** "Find the specific weights that give me 15% return with the *lowest possible risk*."
3.  **Repeat:** We did this for 50 different return levels (from min to max).

### 3. The Result (The White Dashed Line)
The line we plotted connects all these "lowest risk" points.
* Notice how the **Max Sharpe** and **Min Volatility** stars from our simulation sit perfectly (or very close) to this line.
* This confirms that our random simulation was accurate, but the mathematical line is the theoretical limit.

### 4. Summary of the Markowitz Model
* **Input:** Historical Returns & Covariance.
* **Process:** Minimize Variance for a given Target Return.
* **Output:** A set of optimal portfolios (The Efficient Frontier).
* **Lesson:** Diversification is the only "free lunch" in financeâ€”mathematically reducing risk without sacrificing returns.

# ðŸ›¡ï¸ Critical Theory: The Necessity of Backtesting

In quantitative finance, a model without a backtest is just a hypothesis. We cannot assume that a strategy that worked in the past will work in the future. Here is why validation is non-negotiable.

### 1. Why Markowitz *Specifically* Needs Backtesting
The Markowitz model is mathematically elegant but fragile in the real world due to two main issues:
* **"Garbage In, Garbage Out":** The model is an "estimation error maximizer." If your historical data shows a stock had high returns and low risk just by luck, the model will aggressively allocate capital to it.
* **Non-Stationarity:** Financial data changes over time. Correlations that existed in 2020 might break in 2025. If the covariance matrix shifts, your "Optimal Portfolio" can instantly become sub-optimal or dangerous.

### 2. The Universal Rule: The 4 Silent Killers
Every quantitative strategy, from simple moving averages to complex Neural Networks, faces these four hidden risks. Backtesting is the only way to expose them.

#### A. Overfitting (The Hindsight Bias)
* **The Risk:** You can always tune a model's parameters to look perfect on past data.
* **The Reality:** A model that "memorizes" the noise of the past will fail on new, unseen data.

#### B. Transaction Costs & Slippage
* **The Risk:** Theoretical models assume trading is free and instant.
* **The Reality:** Frequent rebalancing (buying/selling to maintain specific weights) incurs fees and spread costs. A strategy earning 15% is useless if it spends 16% on commissions.

#### C. Market Regimes
* **The Risk:** A strategy might perform brilliantly during a Bull Market but collapse during a crisis.
* **The Reality:** You must stress-test strategies across different regimes (e.g., Covid Crash, 2022 Inflation) to ensure robustness.

#### D. Look-Ahead Bias
* **The Risk:** Accidentally using data in your simulation that wouldn't have been available at the time of the trade.
* **Example:** Using the "Close Price" of the day to decide whether to buy at the "Open Price" of the same day.

### Step 6: Backtesting the stratergy

```python

# STEP 6: Backtesting (The Moment of Truth)

# 1. Download "Out-of-Sample" Data (The Test Year: 2025)
test_start = '2025-01-02'
test_end = '2026-01-01'

print(f"Downloading Test Data for {tickers} from {test_start} to {test_end}...")
test_data = yf.download(tickers, start=test_start, end=test_end, auto_adjust=True)['Close']

# 2. Calculate Daily Returns for 2025
test_returns = test_data.pct_change().dropna()

# 3. Construct the Portfolios
# A. The "Optimized" Portfolio (Using weights from Step 3)
# We use dot product (@) to multiply daily returns by our calculated weights
# Portfolio Return = w1*r1 + w2*r2 + ...
optim_daily_returns = test_returns @ max_sharpe_weights

# B. The "Benchmark" Portfolio (Equal Weights)
# This represents a "Lazy" investor who just puts 25% in each stock
equal_weights = np.array([1/len(tickers)] * len(tickers))
benchmark_daily_returns = test_returns @ equal_weights

# 4. Calculate Cumulative Returns (Growth of $1)
# (1 + r) * (1 + r)...
optim_cumulative = (1 + optim_daily_returns).cumprod()
benchmark_cumulative = (1 + benchmark_daily_returns).cumprod()

# 5. Visualizing the Race
plt.figure(figsize=(12, 6))

# Plot Optimized
plt.plot(optim_cumulative, color='lime', linewidth=2, label='Markowitz Max Sharpe (Optimized)')

# Plot Benchmark
plt.plot(benchmark_cumulative, color='white', linestyle='--', linewidth=1.5, alpha=0.7, label='Equal Weight (Naive)')

plt.title('Backtest: 2025 Performance Comparison', fontsize=16)
plt.ylabel('Portfolio Value (Growth of $1)')
plt.xlabel('Date')
plt.legend(fontsize=12)
plt.grid(True, linestyle=':', alpha=0.4)

# Add a final summary text on the chart
total_return_optim = optim_cumulative.iloc[-1] - 1
total_return_bench = benchmark_cumulative.iloc[-1] - 1

plt.text(optim_cumulative.index[0], optim_cumulative.max(), 
         f"Optimized Return: {total_return_optim:.2%}\nBenchmark Return: {total_return_bench:.2%}", 
         fontsize=12, bbox=dict(facecolor='grey', alpha=0.8, edgecolor='white'))

plt.show()

# 6. Conclusion
if total_return_optim > total_return_bench:
    print("âœ… SUCCESS: The Markowitz Optimization beat the naive strategy!")
else:
    print("âŒ INSIGHT: The Optimization underperformed. (Market Regime Shift?)")
```

# ðŸ“‰ Post-Mortem: Why did the Model Fail?

Our backtest revealed that the "Optimized" portfolio underperformed the "Naive" (Equal Weight) benchmark in 2025. This highlights the critical limitations of Mean-Variance Optimization:

### 1. The "Rear-View Mirror" Bias
The Markowitz model assumes the future looks exactly like the past.
* **What happened:** The model optimized for the specific volatility and return patterns of 2020-2024.
* **The Reality:** The market regime shifted in 2025, rendering those historical correlations less predictive.

### 2. Concentration Risk
Markowitz optimizers are often called "Enormous-Error Maximizers."
* If a stock had lucky high returns with low volatility in the past, the model might allocate 50%+ to it.
* If that single stock underperforms in the out-of-sample period, the optimization fails harder than a diversified naive portfolio.

### 3. The Solution (Future Work)
To fix this in a production environment, Quants use:
* **Rolling Rebalancing:** Re-running the optimization every month/quarter rather than holding fixed weights for a year.
* **Shrinkage Estimators:** Mathematically "dampening" the extreme predictions of the covariance matrix (e.g., Ledoit-Wolf).
* **Black-Litterman Model:** Combining historical data with investor "views" to avoid overfitting.
